````python
%%writefile pine_test.cpp
#include <iostream>
using std::cout;
using namespace std;

int main() {
    int niveles = 7; // Número de niveles del pino

    // Loop para cada nivel del pino
    for (int i = 1; i <= niveles; ++i) {
        // Imprime espacios en blanco para alinear el pino
        for (int j = 0; j < niveles - i; ++j) {
            cout << " ";
        }

        // Imprime asteriscos para formar la parte triangular del pino
        for (int k = 0; k < 2 * i - 1; ++k) {
            cout << "*";
        }

        // Salto de línea al final de cada nivel
        cout << endl;
    }

    // Imprime el tronco
    for (int i = 0; i < 2; ++i) {
        // Espacios en blanco para alinear el tronco
        for (int j = 0; j < niveles - 1; ++j) {
            cout << " ";
        }

        // Imprime el tronco
        cout << "|" << endl;
    }

    return 0;
}
````

````python
%%writefile HW4_JorgeG.cpp
#include <iostream>
#include <ctime>

using namespace std;

long long fib(int which) {
    if (which < 2) {
        return which;
    } else {
        return fib(which - 1) + fib(which - 2);
    }
}

int main() {
    int which;
    cout << "Enter the nth Fibonacci number you would like: ";
    cin >> which;

    clock_t start_time, end_time;

    for (int i = 0; i < which; i++) {
        start_time = clock(); // Inicio del tiempo

        long long result = fib(i);

        end_time = clock(); // Fin del tiempo

        double elapsed_time = static_cast<double>(end_time - start_time) / CLOCKS_PER_SEC * 1000; // Convertir a milisegundos

        cout << "The " << i + 1 << "th Fibonacci number is: " << result;
        cout << " and took " << elapsed_time << " milliseconds\n";
    }

    return 0;
}
````

````python
%%writefile HW5_Jorge.cpp
#include<iostream>
#include <ctime>
using namespace std;

int main() {
    int n;
    long long x = 0, y = 1, z = 1;

    cout << "Enter the nth Fibonacci number you would like: ";
    cin >> n;
    if (n < 2) {
        cout << "The " << n << "th Fibonacci number is: " << n << endl;
    } else {
        cout << "The 1st Fibonacci number is: " << x << endl;
        cout << "The 2nd Fibonacci number is: " << y << endl;

        clock_t start_time, end_time;
        for (int i = 3; i <= n; i++) {
            z = x + y;

            cout << "The " << i << "th Fibonacci number is: " << z << " and took ";

            clock_t end_time = clock(); // Fin del tiempo

            double elapsed_time = static_cast<double>(end_time - start_time) / CLOCKS_PER_SEC * 1000; // Convertir a milisegundos

            cout << elapsed_time << " milliseconds\n";

            x = y;
            y = z;
        }
    }
    return 0;
}
````

Una propiedad interesante de los punteros es que pueden utilizarse para acceder directamente a la variable a la que apuntan. Para ello, se antepone al nombre del puntero el operador de desreferencia (*). El propio operador puede leerse como "valor apuntado por".


````python
%%writefile first_pointer.cpp
#include <iostream>
using namespace std;

int main ()
{
  int firstvalue, secondvalue;
  int * mypointer;

  mypointer = &firstvalue;                                                       // to mypointer, le asignamos la dirección de firstvalue
  *mypointer = 10;                                                               // a la dirección de firstvalue, le asignamos el valor 10
  mypointer = &secondvalue;                                                      // to mypointer, le reasignamoso la direccion de secondvalue
  *mypointer = 20;                                                               // a la dirección de secondvalue, le asignamos el valor 20
  cout << "firstvalue is " << firstvalue << '\n';
  cout << "secondvalue is " << secondvalue << '\n';
  return 0;
}
````

````python
%%writefile derived.cpp
#include <iostream>
#include <vector>
#include <cmath>
#include "base.h"

using namespace std;

// Implementación de la función virtual name() para la clase base shape
std::string shape::name() const {
    return "Shape";
}

// Clase derivada rectangle
class rectangle : public shape {
public:
    rectangle(float b, float h) {
        base_ = b;
        height_ = h;
        set_area(calculate_area());
    }

    // Implementación de la función virtual name() para la clase rectangle
    std::string name() const override {
        return "Rectangle";
    }

    float calculate_area() {
        return base_ * height_;
    }

    void set_base(float b) {
        base_ = b;
        set_area(calculate_area());
    }

    void set_height(float h) {
        height_ = h;
        set_area(calculate_area());
    }

private:
    float base_;
    float height_;
};

// Clase derivada circle
class circle : public shape {
public:
    circle(float r) {
        radius_ = r;
        set_area(calculate_area());
    }

    // Implementación de la función virtual name() para la clase circle
    std::string name() const override {
        return "Circle";
    }

    float calculate_area() {
        return 3.14159265358979323846 * radius_ * radius_;
    }

    void set_radius(float r) {
        radius_ = r;
        set_area(calculate_area());
    }

private:
    float radius_;
};

// Clase derivada triangle
class triangle : public shape {
public:
    triangle(float l1, float l2, float l3) {
        if (valid_triangle(l1, l2, l3)) {
            length1 = l1;
            length2 = l2;
            length3 = l3;
            set_area(calculate_area());
        } else {
            set_area(-1);
        }
    }

    // Implementación de la función virtual name() para la clase triangle
    std::string name() const override {
        return "Triangle";
    }

    float calculate_area() {
        float s = (length1 + length2 + length3) / 2;
        return sqrt(s * (s - length1) * (s - length2) * (s - length3));
    }

    void set_lengths(float l1, float l2, float l3) {
        if (valid_triangle(l1, l2, l3)) {
            length1 = l1;
            length2 = l2;
            length3 = l3;
            set_area(calculate_area());
        } else {
            set_area(-1);
        }
    }

private:
    float length1;
    float length2;
    float length3;

    bool valid_triangle(float l1, float l2, float l3) const {
        return (l1 + l2 > l3) && (l1 + l3 > l2) && (l2 + l3 > l1);
    }
};

````

````python
%%writefile HW16_Jorge_try1.cpp
#include <iostream>
#include <vector>
#include <string>

using namespace std;

struct start_and_stop {
    int start_;
    int stop_;
};

vector<start_and_stop> find_transition_points(const string& input_string) {
    vector<start_and_stop> transition_points;
    int n = input_string.size();

    if (n <= 1) return transition_points;

    // Primer enfoque: identificar transiciones iniciales
    int start = 0;
    char last_char = input_string[0];

    for (int i = 1; i < n; ++i) {
        if ((last_char == '0' || last_char == '1') && input_string[i] != last_char) {
            start_and_stop transition;
            transition.start_ = start;
            transition.stop_ = i - 1;
            transition_points.push_back(transition);
            start = i;
        }
        last_char = input_string[i];
    }

    // Segundo enfoque: identificar transiciones adicionales dentro de los bloques
    start = -1;
    last_char = input_string[0];

    for (int i = 1; i < n; ++i) {
        if (last_char != input_string[i] && (last_char == '0' || last_char == '1')) {
            // Si hay una transición entre 0 y 1, registrar la secuencia
            if (start != -1) {
                start_and_stop transition;
                transition.start_ = start;
                transition.stop_ = i - 1;
                transition_points.push_back(transition);
                start = -1; // Reiniciar el marcador de inicio
            }
        } else if (start == -1 && (input_string[i] == '0' || input_string[i] == '1')) {
            // Si encontramos un 0 o 1 después de un bloque de otros números, marcar el inicio de la secuencia
            start = i; // Fijar el inicio como el índice actual
        }
        last_char = input_string[i];
    }

    // Procesar la última transición si el último carácter es 0 o 1
    if (start != -1 && (last_char == '0' || last_char == '1')) {
        start_and_stop last_transition;
        last_transition.start_ = start;
        last_transition.stop_ = n - 1;
        transition_points.push_back(last_transition);
    }

    return transition_points;
}


int main() {
    string input_string = "00001110293411111887888880000222333311111";

    vector<start_and_stop> transition_points = find_transition_points(input_string);

    cout << "start   stop" << endl;
    for (int i = 0; i < transition_points.size(); ++i) {
        cout << "  " << transition_points[i].start_ << "    " << transition_points[i].stop_ << endl;
    }

    return 0;
}
````
